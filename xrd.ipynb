# Tâche 7: Diffractogramme

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
```

```python
with MPRester("Lyz7DTq5tVKRlBe9") as m:
    structure = m.get_structure_by_material_id("mp-672285") 
```

```python
Lambda = 1.54060 #un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å.
pattern = XRDCalculator(Lambda).get_pattern(structure) # Pattern de diffraction associé au cristal (structure) pour un certain Lambda
```

```python
def to_print(num):
    print("pic ", num+1, " pour un rayonnement X de longueur d’onde: λ=",Lambda, "Å :")
    print("        Angle 2Θ :", pattern.x[num],"°")
    print("        Intensité du pic :", pattern.y[num])
    print("        Indices hkl et multiplicité :", pattern.hkls[num][0])
    print("        d :", pattern.d_hkls[num], "\n")
```

# Déterminer les 3 premiers pics du diffractogramme

```python
for i in range(3):
    to_print(i)
```
